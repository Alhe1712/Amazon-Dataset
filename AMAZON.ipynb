```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

### ANALISIS EXPLORATORIO DE DATOS


```python
amazon=pd.read_csv('amazon.csv')
amazon.head(5)
```

```python
amazon.info()
```

```python
amazon.isna().sum()
```

```python
amazon.duplicated().sum()
```

```python
amazon.dropna(inplace=True)
```

```python
# Convertir las columnas de precios y descuentos a tipo num√©rico
amazon['discounted_price'] = amazon['discounted_price'].str.replace('‚Çπ', '', regex=False).str.replace(',', '').astype(float)
amazon['actual_price'] = amazon['actual_price'].str.replace('‚Çπ', '', regex=False).str.replace(',', '').astype(float)
amazon['discount_percentage'] = amazon['discount_percentage'].str.replace('%', '', regex=False).astype(float)
amazon['rating'] = amazon['rating'].str.replace('|', '', regex=False)
amazon['rating'] = pd.to_numeric(amazon['rating'], errors='coerce')
amazon['rating_count'] = amazon['rating_count'].str.replace(',', '').astype(int)
```

```python
print(amazon.head(10))
```

### DESCUENTOS

```python
#Calcular la correlaci√≥n entre discount_percentage y rating
correlation = amazon['discount_percentage'].corr(amazon['rating'])

# Imprimir el resultado
print(f"La correlaci√≥n entre el descuento y la calificaci√≥n es: {correlation:.2f}")
```

 -0.16, lo que indica una relaci√≥n muy d√©bil y negativa entre el descuento y las calificaciones de los productos. Esto sugiere que no hay una relaci√≥n clara entre un mayor descuento y una mejor o peor calificaci√≥n en este caso. En otras palabras, el descuento por s√≠ solo no parece influir significativamente en la calificaci√≥n de los productos

```python
# C√°lculo del descuento absoluto y porcentaje real
amazon['discount_amount'] = amazon['actual_price'] - amazon['discounted_price']
amazon['real_discount_percent'] = (amazon['discount_amount'] / amazon['actual_price']) * 100
amazon['real_discount_percent'] = amazon['real_discount_percent'].round(2)
```

```python
sample = amazon.sample(30, random_state=1)  # Muestra de 30 productos para que el gr√°fico no est√© saturado

plt.figure(figsize=(12, 4))
plt.plot(sample['actual_price'].values, label='Precio original', marker='o')
plt.plot(sample['discounted_price'].values, label='Precio con descuento', marker='o')
plt.title('Comparaci√≥n de precios originales vs con descuento')
plt.xlabel('Producto (muestra aleatoria)')
plt.ylabel('Precio')
plt.legend()
plt.grid(False)
plt.show()
```

Los productos con mayor precio original tienden a tener los descuentos m√°s grandes (en t√©rminos absolutos).
En productos baratos, el precio con descuento a veces se acerca mucho al original, lo cual puede indicar un descuento peque√±o.
El descuento no es uniforme: var√≠a dependiendo del producto.

```python
top_discounted = amazon.sort_values('real_discount_percent', ascending=False)
top_discounted = top_discounted.drop_duplicates(subset='product_name').head(10)
top_discounted= top_discounted[['product_name', 'actual_price', 'discounted_price', 'real_discount_percent']]

top_discounted['product_short_name'] = top_discounted['product_name'].str.slice(0, 25) +'...' # Limitar el nombre del producto a 25 caracteres
plt.figure(figsize=(12, 4))
sns.barplot(data=top_discounted, x='product_short_name', y='real_discount_percent', palette='coolwarm', hue='product_short_name')
plt.xticks(rotation=45, ha='right')
plt.yticks(np.arange(50, 110, 10))
plt.ylim(50, top_discounted['real_discount_percent'].max() + 10)
plt.xlabel('Producto')
plt.ylabel('Descuento real (%)')  
plt.title('Top 10 productos con mayor descuento real')
plt.show()
```

```python

plt.figure(figsize=(6, 4))
sns.scatterplot(data=amazon, x='discount_amount', y='real_discount_percent')
plt.title('Relaci√≥n entre descuento absoluto y porcentaje de descuento')
plt.xlabel('Descuento en ‚Çπ')
plt.ylabel('Descuento %')
plt.grid(True)
plt.show()
```

Observaciones: Hay una clara dispersi√≥n: no hay una relaci√≥n lineal fuerte entre el descuento en ‚Çπ y el porcentaje de descuento.
La mayor√≠a de los puntos tienen descuentos absolutos bajos (entre 0 y 5000).
Pero esos descuentos a√∫n pueden representar un porcentaje alto, especialmente en productos baratos.
Muchos productos tienen descuentos entre 20% y 70%.
Algunos puntos se acercan al 90% de descuento.
Productos con descuentos altos en dinero, pero no tan altos en porcentaje:
En la parte derecha del gr√°fico (descuentos absolutos altos, >20,000), los porcentajes tienden a ser m√°s bajos (30%‚Äì50%). Esto sugiere que son productos muy caros, donde incluso un gran descuento en dinero representa una proporci√≥n m√°s peque√±a del precio original.

No todos los productos tienen una pol√≠tica de descuentos uniforme, lo cual puede ser √∫til para analizar estrategias de precios o promociones.



```python
# Calcular descuentos promedio por categor√≠a
category_discounts = amazon.groupby('category')['real_discount_percent'].mean().sort_values(ascending=False)

# Seleccionar el top 10
top_discounts = category_discounts.head(10)

# Gr√°fico
plt.figure(figsize=(10, 6))
bars = plt.barh(top_discounts.index, top_discounts.values, color='skyblue')
plt.title('Top 10 categor√≠as con mayor descuento promedio (%)')
plt.xlabel('Descuento promedio (%)')
plt.gca().invert_yaxis()  

for bar in bars:
    width = bar.get_width()
    plt.text(width + 1, bar.get_y() + bar.get_height()/2,
             f'{width:.2f}%', va='center')

plt.tight_layout()
plt.show()
```

Observaciones: 
La mayor√≠a de las categor√≠as est√°n en el rubro de tecnolog√≠a y accesorios. Tres categor√≠as tienen un descuento promedio alt√≠simo (cerca del 90%):

Earpads, 
Phone charms
Cable protectors

Esto sugiere que hay sobreoferta, baja rotaci√≥n, o son productos usados para atraer tr√°fico (ofertas gancho).

Rango de descuentos promedio:
Va de 76% a 90%, lo cual es muy alto.
Incluso el d√©cimo lugar (76.15%) sigue siendo una rebaja considerable.

Si eres consumidor: estas categor√≠as son excelentes para encontrar verdaderas ofertas.
Si eres vendedor o analista: estas categor√≠as probablemente tienen altos m√°rgenes iniciales o poca demanda, permitiendo aplicar descuentos agresivos sin p√©rdidas o para limpiar inventario.



### VENTAS POR CATEGORIA

```python
# Calcular las ventas estimadas
amazon['estimated_sales_billions'] = (amazon['discounted_price'] * amazon['rating_count'])/1e9

sales_by_category = amazon.groupby('category')['estimated_sales_billions'].sum().reset_index().sort_values(by='estimated_sales_billions', ascending=False)
print(sales_by_category.head(10))
```

```python
# Gr√°fico de barras horizontales para las categor√≠as con mayores ventas en billones
df_plot = sales_by_category.copy()
top_sales = df_plot.sort_values(by='estimated_sales_billions', ascending=False).head(10)

plt.figure(figsize=(14, 6))
sns.barplot(
    data=top_sales,
    x='estimated_sales_billions',
    y='category',
    palette='viridis',
    hue='category'
)

plt.title('Top 10 categor√≠as con mayores ventas', fontsize=14)
plt.xlabel('Ventas estimadas (miles de millones)', fontsize=12)
plt.ylabel('Categor√≠a', fontsize=12)
plt.tight_layout()
plt.show()
```

```python
# Gr√°fico de barras horizontales para las categor√≠as con menores ventas en millones
df_plot['estimated_sales_thousands'] = df_plot['estimated_sales_billions'] * 1_000_000
low_sales = df_plot.sort_values(by='estimated_sales_thousands', ascending=True).head(10)

plt.figure(figsize=(14, 6))
sns.barplot(
    data=low_sales,
    x='estimated_sales_thousands',
    y='category',
    palette='viridis',
    hue='category'
)

plt.title('Top 10 categor√≠as ventas bajas', fontsize=14)
plt.xlabel('Ventas estimadas (millones)', fontsize=12)
plt.ylabel('Categor√≠a', fontsize=12)
plt.tight_layout()
plt.show()
```

### RATING DE PRODUCTOS

```python
#productos con mejor calificaci√≥n
top_ratings = amazon.sort_values('rating', ascending=False).head(10)
top_ratings = top_ratings[['product_name', 'product_id', 'rating']]
print(top_ratings)
```

```python
# productos con peor calificaci√≥n
low_ratings = amazon.sort_values('rating', ascending=True).head(10)
low_ratings = low_ratings[['product_name', 'product_id', 'rating']]
print(low_ratings)
```

```python
fig, axes = plt.subplots(2, 1, figsize=(12, 10))

# A√±adir una columna corta para los nombres de los productos
top_ratings['short_name'] = top_ratings['product_name'].str.slice(0, 25) + '...'
low_ratings['short_name'] = low_ratings['product_name'].str.slice(0, 25) + '...'
# Para top_ratings
x_top = range(len(top_ratings))
axes[0].bar(x_top, top_ratings['rating'], color='purple', alpha=0.4)
axes[0].set_title('Top 10 productos con mejor calificaci√≥n')
axes[0].set_ylabel('Calificaci√≥n')
axes[0].set_yticks(np.arange(0, 6.5, 0.5))
axes[0].set_xticks(x_top)
axes[0].set_xticklabels(top_ratings['short_name'], rotation=45, ha='right')

# Para low_ratings
x_low = range(len(low_ratings))
axes[1].bar(x_low, low_ratings['rating'], color='tomato', alpha=0.6)
axes[1].set_title('Top 10 productos con peor calificaci√≥n')
axes[1].set_xlabel('Producto')
axes[1].set_ylabel('Calificaci√≥n')
axes[1].set_yticks(np.arange(0, 6.5, 0.5))
axes[1].set_xticks(x_low)
axes[1].set_xticklabels(low_ratings['short_name'], rotation=45, ha='right')

plt.tight_layout()
plt.show()

```
